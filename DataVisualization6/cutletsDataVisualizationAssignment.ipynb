```python
# A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.

# H0: There is no significant difference in the diameter of the cutlet between two units
# H1: There is significant difference in the diameter of the cutlet between two units
```

```python
# Assumptions:
# 1. The data is normally distributed
# 2. The data is independent 
# 3. The data is homoscedastic (having equal variance)
```

```python
# Test:
# 1. Shapiro-Wilk Test
# 2. Bartlett's Test
```

```python
# Shapiro-Wilk Test
from scipy.stats import shapiro
 
```

```python
# for Unit A we check
stat, p = shapiro(df['Unit A'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')  
 
```

```python
# for unit B we check
stat, p = shapiro(df['Unit B'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

```python
# Bartlett's Test
from scipy.stats import bartlett
stat, p = bartlett(df['Unit A'], df['Unit B'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Sample looks homoscedastic (fail to reject H0)')
else:
    print('Sample does not look homoscedastic (reject H0)')
```

```python
# Since the data is normally distributed and homoscedastic.
#  we can use two sample t-test to check the significance of the difference in the diameter of the cutlet between two units.
```

```python
# Two Sample T-Test
from scipy.stats import ttest_ind
stat, p = ttest_ind(df['Unit A'], df['Unit B'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')

 
```

```python
#  since p value is greater than 0.05, we fail to reject the null hypothesis.
#  there is no significant difference in the diameter of the cutlet between two units.
 
```
