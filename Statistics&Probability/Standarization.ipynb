<h1>Why do we need to standardize data in Python?</h1>
<p>
Before diving deep into the concept of standardization, it is very important for us to know the need for it.</p>
<p>
So, you see, the datasets which we use to build a model for a particular problem statement is usually built from various sources. Thus, it can be assumed that the data set contains variables/features of different scales.</p>

<h2>Standardization</h2>
<p><b>Standarization</b> is used on the data values that are normally distribute. Further, by applying standardization, we tend to make mean of the dataset as 0 and the standard deviation equivlent to 1.</p>
<br>
<p>That is, by standardizing the values, we get the following statistics of the
data distribution<br>
mean=0<br>
standard deviation =1</p>
